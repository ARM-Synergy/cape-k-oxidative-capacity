```python
import os
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import matplotlib.cm as cm  # For colormap

# Define the directory containing the trajectory files
tdump_dir = r"C:\Users\smkasimongwa\OneDrive - The University of Melbourne\Desktop\D01\D05"

# Function to parse a HYSPLIT trajectory file and extract only Layer 1 (10m AGL)
def parse_hysplit_layer_1(filepath):
    latitudes = []
    longitudes = []

    with open(filepath, 'r') as file:
        lines = file.readlines()
        for line in lines:
            parts = line.split()
            try:
                traj_id = int(parts[0])  # First column is the trajectory ID (layer)
                if traj_id == 1:  # Filter only Layer 1 (10 AGL)
                    latitude = float(parts[9])  # 10th column
                    longitude = float(parts[10])  # 11th column
                    latitudes.append(latitude)
                    longitudes.append(longitude)
            except (IndexError, ValueError):
                continue  # Skip lines that do not contain trajectory data

    return latitudes, longitudes

# Get a list of all `.txt` files in the directory
tdump_files = sorted([os.path.join(tdump_dir, f) for f in os.listdir(tdump_dir) if f.endswith('.txt')])

# Generate distinct colors for each file (day)
file_colors = cm.viridis([i / len(tdump_files) for i in range(len(tdump_files))])

# Initialize the map focusing on Tasmania and surrounding ocean
plt.figure(figsize=(12, 10))
m = Basemap(projection='merc',
            llcrnrlat=-47,  # Lower boundary (Southern Ocean)
            urcrnrlat=-38,  # Upper boundary (just above Tasmania)
            llcrnrlon=140,  # Lower-left longitude (Bass Strait)
            urcrnrlon=155,  # Upper-right longitude (Tasman Sea)
            resolution='i')  # Intermediate resolution

# Add shaded relief for topography
m.shadedrelief()

# Add coastlines and country boundaries
m.drawcoastlines()
m.drawcountries()

# Add parallels and meridians
m.drawparallels(range(-47, -37, 2), labels=[1, 0, 0, 0])  # Latitude lines every 2 degrees
m.drawmeridians(range(140, 156, 2), labels=[0, 0, 0, 1])  # Longitude lines every 2 degrees

# Plot only Layer 1 (10 AGL) with unique colors per day
for file_idx, tdump_file in enumerate(tdump_files):
    latitudes, longitudes = parse_hysplit_layer_1(tdump_file)

    if latitudes and longitudes:
        x, y = m(longitudes, latitudes)
        file_date = os.path.basename(tdump_file).replace('.txt', '')  # Extract date from filename
        m.plot(x, y, linewidth=2, color=file_colors[file_idx], marker='o', markersize=3, label=f"{file_date}")

# Add legend with dates
plt.legend(fontsize=10, loc='upper right', title="Trajectory Date")

# Add title
plt.title('Cape-Grim-HYSPLIT Trajectories (2024-11-24)')
plt.show()

print("Trajectory visualization complete with Tasmania-focused map!")
```

```python
import os
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import matplotlib.cm as cm  # For colormap

# Define directories for each date
tdump_dirs = {
    "2024-11-23": r"C:\Users\smkasimongwa\OneDrive - The University of Melbourne\Desktop\D01\D06",
    "2024-11-24": r"C:\Users\smkasimongwa\OneDrive - The University of Melbourne\Desktop\D01\D05",
    "2024-11-27": r"C:\Users\smkasimongwa\OneDrive - The University of Melbourne\Desktop\D01\D08",
    "2024-11-28": r"C:\Users\smkasimongwa\OneDrive - The University of Melbourne\Desktop\D01\D07"
}

# Function to parse a HYSPLIT trajectory file and extract only Layer 1 (10m AGL)
def parse_hysplit_layer_1(filepath):
    latitudes, longitudes = [], []
    with open(filepath, 'r') as file:
        lines = file.readlines()
        for line in lines:
            parts = line.split()
            try:
                traj_id = int(parts[0])  # First column is the trajectory ID (layer)
                if traj_id == 1:  # Filter only Layer 1 (10 AGL)
                    latitude = float(parts[9])  # 10th column
                    longitude = float(parts[10])  # 11th column
                    latitudes.append(latitude)
                    longitudes.append(longitude)
            except (IndexError, ValueError):
                continue  # Skip lines that do not contain trajectory data
    return latitudes, longitudes

# Create 2x2 subplot layout with slightly larger figure size
fig, axes = plt.subplots(2, 2, figsize=(15, 12))  
fig.suptitle("Cape Grim - Backtrajectory", fontsize=16, fontweight='bold')

# Generate distinct colors for each hour (00 to 23)
hourly_colors = cm.viridis([i / 23 for i in range(24)])

# Dictionary to store legend entries (avoid repetition)
legend_labels = set()
legend_entries = []

# Iterate through each subplot position and date
for ax, (date, tdump_dir) in zip(axes.flat, tdump_dirs.items()):
    # Get the list of `.txt` files in the directory
    tdump_files = sorted([os.path.join(tdump_dir, f) for f in os.listdir(tdump_dir) if f.endswith('.txt')])
    
    # Initialize the map for each subplot
    m = Basemap(projection='merc',
                llcrnrlat=-47, urcrnrlat=-38,  # Latitude bounds
                llcrnrlon=140, urcrnrlon=155,  # Longitude bounds
                resolution='i', ax=ax)  # Pass subplot axis

    # Add features
    m.shadedrelief()
    m.drawcoastlines()
    m.drawcountries()
    m.drawparallels(range(-47, -37, 2), labels=[1, 0, 0, 0])
    m.drawmeridians(range(140, 156, 2), labels=[0, 0, 0, 1])

    # Plot trajectories grouped by hour
    for file_idx, tdump_file in enumerate(tdump_files):
        latitudes, longitudes = parse_hysplit_layer_1(tdump_file)
        if latitudes and longitudes:
            x, y = m(longitudes, latitudes)
            hour = int(os.path.basename(tdump_file).split("_")[-1].split(".")[0])  # Extract hour
            
            # Plot with hour-based color
            plot_line, = m.plot(x, y, linewidth=2, color=hourly_colors[hour], marker='o', markersize=3, label=f"{hour:02d}:00")

            # Store legend entries once to avoid duplicates
            if f"{hour:02d}:00" not in legend_labels:
                legend_entries.append(plot_line)
                legend_labels.add(f"{hour:02d}:00")

    # Add subplot title
    ax.set_title(date, fontsize=16, fontweight='bold')

# Move legend closer and increase font size to 16
fig.legend(handles=legend_entries, loc='center right', fontsize=16, title="Hours", title_fontsize=16, bbox_to_anchor=(1.02, 0.5), borderaxespad=0.)

# Adjust layout to fit legend closer
plt.tight_layout(rect=[0, 0, 0.92, 0.96])  # Move figure left slightly
plt.show()

print("Trajectory visualization complete with a 2x2 layout, closer legend, and increased font size!")
```
